```python
#Load the CSV file using pandas.
df=pd.read_csv('sales2.csv')
```

```python
#Show column names.
df.columns
```

```python
#Get all employees from IT department.
df[df["Department"]=="IT"]
```

```python
#Employees whose salary > 40000.
df[df["Salary"]>40000]
```

```python
#Employees from Chennai.
df[df["City"]=="Chennai"]
```

```python
#Employees with experience >= 5 years.
df[df["Experience"]>=5]
```

```python
#HR employees with salary < 35000.
df[(df["Department"]=="HR") & (df["Salary"]<35000)]
```

```python
#Find average salary.
df["Salary"].mean()
```

```python
#Find maximum salary.
df[["Salary"]].max()
```

```python
#Find minimum age.
df[["Age"]].min()
```

```python
#Average salary by department.
df.groupby("Department")["Salary"].mean().reset_index()
```

```python
#Count employees in each city.
df.groupby("City")["EmpID"].count().reset_index()
```

```python
#Sort by salary (ascending).
df.sort_values("Salary",ascending=True)
```

```python
#Sort by age (descending).
df.sort_values("Age",ascending=False)
```

```python
#Sort by experience then salary.
df.sort_values(by=["Experience","Salary"],ascending=[True,True])
```

```python
#Department-wise average salary.
df.groupby("Department")["Salary"].mean().reset_index()
```

```python
#City-wise employee count.
df.groupby("City")["EmpID"].count().reset_index()
```

```python
#Department-wise maximum experience.
df.groupby("Department")["Experience"].max().reset_index()
```

```python
#Average salary of IT employees per city.
df[df["Department"]=="IT"].groupby("City")["Salary"].mean().reset_index()
```

```python
#Add a new column Bonus = Salary * 0.10.
df["Bonus"]=df["Salary"]*0.10
df
```

```python
#Employees earning more than department average salary.
df[df["Salary"]>df.groupby("Department")["Salary"].transform("mean")]
```

```python
#Top 5 highest paid employees.
df.sort_values("Salary",ascending=False).head(5)
```

```python
#Employees with experience between 5 and 10.
df[df["Experience"].between(5,10)]
```

```python
#Rename column EmpID to Employee_ID.
df.rename(columns={"EmpID":"Employee_ID"})
```

```python
df.drop(columns="City")
```
